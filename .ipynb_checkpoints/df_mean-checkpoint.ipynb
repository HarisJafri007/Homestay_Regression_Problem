This Data Frame is a transformed version of Original Data Frame with Null Values replaced with Mean and we have not treated outliers!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style
```

```python
file_path = 'df_mean.csv'
df = pd.read_csv(file_path)
df
```

Let us watch the Correlation between different features!

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

Now we are going to seperate our data frame into two parts. X containing all Independent Variables and y containing the Target Variable.

```python
X = df.drop('log_price',axis=1)
```

```python
X.head()
```

```python
y = df[['log_price']]
```

```python
y.head()
```

We have successfully divided the Data Frame into Independent and Target Variables. Now we shall divide the Data Frame into Test and Train!

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train
```

```python
X_test
```

This clearly shows that our model has been successfully divided into Train Data and Test Data!

```python
round(54117*0.3)
```

Let us import some Machine Learning Libraries!

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

### Linear Regression

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
# Predicting y values using the test set
y_predicted = regression_model.predict(X_test)

# Ensure y_test and y_predicted are arrays
y_test = y_test.values.flatten()
y_predicted = y_predicted.flatten()

# Comparing y_predicted to y_test
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_predicted})
print(comparison.head())
```

```python
print(comparison.head(10))
```

```python
regression_model.score(X_train,y_train) #Accuracy
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_predicted)) #Average Magnitude of Individual Errors
```

### Decision Tree

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dct_model = DecisionTreeRegressor(random_state=1)
```

```python
dct_model.fit(X_train,y_train)
```

```python
# Predicting y values using the test set
y_predicted = dct_model.predict(X_test)

# Ensure y_test and y_predicted are arrays
y_test = y_test.values.flatten()
y_predicted = y_predicted.flatten()

# Comparing y_predicted to y_test
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_predicted})
print(comparison.head())
```

```python
dct_model.score(X_train, y_train) #Accuracy, how will the Regression Line fits the Data Points
```

```python
print(mean_absolute_error(y_test,y_predicted)) # Mean Absolute Error
```
